# NAIVE BYES 

What is Naive Bayes?

Naive Bayes is a Supervised Machine Learning algorithm mainly used for:

Classification problems,
Text classification,
Spam detection,
Sentiment analysis,
Medical diagnosis,
Recommendation systems

It is based on Bayes Theorem of probability (CONDITIONAL PROBABILITY)

# Types of Naive Bayes

# Gaussian Naive Bayes

Used when data is:

Continuous
Normally distributed

Example:

Height
Weight
Salary

Mostly numerical datasets

work based on 
```math
f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}

```

# Multinomial Naive Bayes

Used for:

Text data
Word counts
NLP

Example:

Spam detection,
News classification

Works on frequency/counts 

# Bernoulli Naive Bayes

Used for:

Binary features
Yes/No type data

Example:

Word present or absent

# DIFFERNCE B/W GAUSSIANB , BERNOULLIINB , MULTINOMIALNB

| Feature         | GaussianNB     | BernoulliNB | MultinomialNB   |
| --------------- | -------------- | ----------- | --------------- |
| Data Type       | Continuous     | Binary      | Count/Frequency |
| Example         | Height, Weight | Yes/No      | Word counts     |
| Values          | 12.5, 170      | 0 or 1      | 0,1,2,3         |
| Best For        | Numerical data | Binary text | NLP/Text        |
| Uses Frequency? | No             | No          | Yes             |




```TEXT
Gaussian → Continuous numbers
Bernoulli → Presence/Absence
Multinomial → Frequency/Counts
```
