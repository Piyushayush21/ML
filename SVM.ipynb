# What is SVM (Support Vector Machine)

Definition

SVM (Support Vector Machine) is a supervised machine learning algorithm used for:

Classification
Regression

Find the best boundary (hyperplane)
that separates different classes

# Important Terms in SVM 
| Term            | Meaning                                       |
| --------------- | --------------------------------------------- |
| Hyperplane      | Decision boundary                             |
| Margin          | Distance between boundary and nearest points  |
| Support Vectors | Closest data points to hyperplane             |
| Kernel          | Converts nonlinear data into higher dimension |


# Hyperplane
Definition

A hyperplane is the decision boundary that separates classes

Hyperplane Equation

w⋅x+b=0

Where:

w = weights

x = input features

b = bias

# Margin

Definition

Margin is the distance between:

hyperplane
nearest data points

SVM tries to maximize this margin

large margin means:

better generalization
less overfitting

# Support Vectors
Definition

Support vectors are the nearest data points to the hyperplane.

These points decide:

position of boundary

# Kernel
Definition

A kernel is a mathematical function used in SVM to convert:
low dimensional data into high demensioanal data so that in can be easily sepreable 

# kernal comparison 
| Kernel     | Best For                     |
| ---------- | ---------------------------- |
| Linear     | Linear data                  |
| Polynomial | Curved relationships         |
| RBF        | Complex nonlinear data       |
| Sigmoid    | Neural-network-like problems |

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error , accuracy_score , r2_score
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
```

```python
data = fetch_california_housing()
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df.head
```

```python
df['price'] = data.target
df.head()
```

```python
x = df.drop("price", axis = 1)
y = df["price"]
```

```python
x
```

```python
y 
```

```python
x_train ,x_test , y_train , y_test = train_test_split(x,y , test_size=0.2 , random_state= 42)
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = SVR(kernel='linear')
```

```python
model.fit(x_train_scaled , y_train)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
y_pred = model.predict(x_test_scaled)
y_pred
```

```python
x.shape
```

```python
mse = mean_squared_error(y_test, y_pred)
print("MSE =", mse)
```

```python
r2 = r2_score(y_test,y_pred)
r2
```
