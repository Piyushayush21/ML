# Hyperparameter

A hyperparameter is a configuration variable whose value is set before the learning process begins and controls the behavior, structure, or training process of a machine learning algorithm.

Hyperparameter tuning is the process of selecting the optimal set of hyperparameter values for a machine learning model in order to maximize model performance on unseen data.

Examples:

K in KNN

C in SVM

# 1. KNN
n_neighbors

Number of nearest training samples considered during classification or prediction.

# 2. SVC
C Parameter:
Regularization parameter that controls the trade-off between maximizing margin and minimizing classification error.

Gamma:
Defines the influence radius of a single training sample in nonlinear kernels.

Kernel:
Mathematical function used to transform data into higher-dimensional feature space for linear separability.

# 4. Naive Bayes
Alpha:
Smoothing parameter used to avoid zero probability problems in probabilistic classification.

Technical Definition of Grid Search:

Grid Search is an exhaustive hyperparameter optimization technique that evaluates all possible combinations of specified hyperparameter values using cross-validation.

Technical Definition of Random Search:

Random Search is a stochastic hyperparameter optimization method that randomly samples combinations of hyperparameter values from a predefined search space.

Technical Definition of Cross Validation:

Cross-validation is a resampling technique used to evaluate model performance by partitioning the dataset into multiple training and validation subsets.

```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
a = load_breast_cancer()
x = a.data
y = a.target

print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
knn = KNeighborsClassifier()
```

```python
knn_params = {
    'n_neighbors': [3,5,7,9,11],
    'metric': ['euclidean', 'manhattan'],
    'weights': ['uniform', 'distance']
}
```

```python
knn_grid = GridSearchCV(
    estimator=knn,
    param_grid=knn_params,
    cv=5,
    scoring='accuracy'
)

knn_grid.fit(x_train, y_train)
```

```python
best_knn = knn_grid.best_estimator_

knn_pred = best_knn.predict(x_test)

print("KNN Accuracy:")
print(accuracy_score(y_test, knn_pred))
```

```python
svc = SVC()
```

```python
svc_params = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
```

```python
svc_grid = GridSearchCV(
    estimator=svc,
    param_grid=svc_params,
    cv=5,
    scoring='accuracy'
)

svc_grid.fit(x_train, y_train)
```

```python
best_svc = svc_grid.best_estimator_

svc_pred = best_svc.predict(x_test)

print("SVC Accuracy:")
print(accuracy_score(y_test, svc_pred))
```

```python
nb = GaussianNB()
```

```python
params = {
    'var_smoothing': np.logspace(0, -9, num=100)
}
```

```python
grid = GridSearchCV(
    estimator=nb,
    param_grid=params,
    cv=5,
    scoring='accuracy'
)
```

```python
grid.fit(x_train, y_train)
```

```python
print("Best Parameter:")
print(grid.best_params_)
```

```python
print("Best Cross Validation Accuracy:")
print(grid.best_score_)
```

```python
best_nb = grid.best_estimator_
```

```python
y_pred = best_nb.predict(x_test)
```

```python
print("KNN Accuracy:", accuracy_score(y_test, knn_pred))
print("SVC Accuracy:", accuracy_score(y_test, svc_pred))
print("Naive Bayes Accuracy:", accuracy_score(y_test, y_pred))
```
