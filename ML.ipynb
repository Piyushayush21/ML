# NORMALIZATION 

 #### Normalization in Machine Learning is a feature scaling technique used to transform numerical data into a common scale or range without distorting the differences in the values. It ensures that all features contribute proportionally to the learning process of the model.

 #### The most commonly used normalization method is Min-Max Scaling, where values are transformed into a fixed range, usually between 0 and 1.

Normalized Value =
                    (Current Value − Minimum Value)/(Maximum Value − Minimum Value)

####  Why Normalization is Needed
Normalization helps:
Faster training
Better accuracy
Gradient descent stable hota hai
Distance-based algorithms better work karte hain

Especially important in:

KNN
K-Means
Logistic Regression
Neural Networks
SVM

````python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd 
data = {

    "marks" : [40, 50, 60 , 80,100]
}
df = pd.DataFrame(data)
scalar = MinMaxScaler()
df['Normalized'] = scalar.fit_transform(df[["marks"]])
````

````python
print(df)
````

# STANDARDIZATION 

standardization is a feature scaling technique used in Machine Learning to transform data so that:
mean becomes 0
standard deviation becomes 1
it scales data around the center (mean) while preserving the distribution pattern

also called z-score normalization 

Standardized Value =
(Current Value − Mean)
÷
Standard Deviation

```text
Standardization Helps In:
Features ko balanced scale par laata hai
Model training faster hoti hai
Gradient descent efficiently converge karta hai
Large value features domination nahi karte
Gaussian distributed data ke liye better work karta hai
Optimization stable hoti hai
Especially Important In
Linear Regression
Logistic Regression
Support Vector Machine (SVM)
Principal Component Analysis (PCA)
Neural Networks

```

````python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
data = { 
      'Age': [18, 22, 25, 30, 35],
    'Salary': [20000, 30000, 50000, 80000, 150000]
    }
df = pd.DataFrame(data)
df
````

````python
scalar = StandardScaler()
scaled_data = scalar.fit_transform(df)
````

````python
print(scaled_data)
````

# Differnce between normalization and standardization 

| Normalization                                   | Standardization                                        |
| ----------------------------------------------- | ------------------------------------------------------ |
| Scales data into a fixed range (usually 0 to 1) | Transforms data so mean = 0 and standard deviation = 1 |
| Uses minimum and maximum values                 | Uses mean and standard deviation                       |
| Also called Min-Max Scaling                     | Also called Z-score Scaling                            |
| Sensitive to outliers                           | Less sensitive to outliers                             |
| Best when data distribution is not Gaussian     | Best when data follows Gaussian/normal distribution    |
