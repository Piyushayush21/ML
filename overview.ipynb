# Linear Regression

Definition:

Linear Regression is a regression algorithm used to predict continuous numerical values when the relationship between input and output is linear.

It tries to fit a straight line between:

independent variable (X)
dependent variable (y)

```text
Dataset
   ↓
Train-Test Split
   ↓
Create LinearRegression Model
   ↓
Train Model
   ↓
Prediction
   ↓
Evaluate Model
```

## Multiple Linear Regression

Definition:

Multiple Linear Regression predicts a continuous value using multiple input features instead of one.

Example:

house price using:
area,
bedrooms,
location,

## Polynomial Regression

Definition

Polynomial Regression is used when the relationship between input and output is nonlinear or curved.

Instead of straight line:

it fits a curve.


```text
Dataset
   ↓
Convert Features to Polynomial Features
   ↓
Train-Test Split
   ↓
Train Model
   ↓
Prediction
   ↓
Evaluation
```

## SVR (Support Vector Regression)
Definition

SVR is the regression version of Support Vector Machine (SVM).

It predicts continuous values while maintaining:

maximum margin
minimum error

Works well for:

nonlinear data
complex datasets

```text
Dataset
   ↓
Feature Scaling
   ↓
Train-Test Split
   ↓
Select Kernel
   ↓
Train SVR Model
   ↓
Prediction
   ↓
Evaluation
```
