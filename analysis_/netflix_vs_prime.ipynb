```python
netflix_genres = netflix['listed_in'].str.split(',')

all_genres = []

for genres in netflix_genres.dropna():
    for genre in genres:
        all_genres.append(genre.strip())

genre_counts = pd.Series(all_genres).value_counts()

print(genre_counts.head(10))
```

```python
genre_counts.head(10).plot(kind='bar')

plt.title("Top Netflix Genres")
plt.show()
```

```python
netflix_subscribers = 300000000

monthly_price = 649

monthly_revenue = netflix_subscribers * monthly_price

print(monthly_revenue)
```

```python
years = 5

subs = [300000000]

for i in range(years):
    
    next_year = subs[-1] * 0.95
    
    subs.append(next_year)

print(subs)
```

```python
revenue = []

for s in subs:
    
    revenue.append(
        s * 649 * 12
    )

forecast = pd.DataFrame({
    "Year": range(0,6),
    "Subscribers": subs,
    "Revenue": revenue
})

print(forecast)
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    forecast['Year'],
    forecast['Revenue'],
    marker='o'
)

plt.title("Netflix Revenue Forecast")

plt.xlabel("Year")

plt.ylabel("Revenue")

plt.show()
```

```python
simulations = 1000

years = 5

results = []

for sim in range(simulations):

    subscribers = 300000000

    for year in range(years):

        churn = np.random.uniform(
            0.02,
            0.08
        )

        subscribers *= (1-churn)

    results.append(subscribers)
```

```python
plt.figure(figsize=(10,5))

plt.hist(
    results,
    bins=30
)

plt.title(
    "Monte Carlo Subscriber Forecast"
)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample business data
data = {
    "Company": ["Netflix", "Amazon Prime"],
    "Subscribers_Million": [302, 220],
    "Revenue_Billion_USD": [39, 638]
}

df = pd.DataFrame(data)

print(df)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Company',
    y='Subscribers_Million',
    data=df
)

plt.title("Netflix vs Amazon Prime Subscribers")
plt.ylabel("Subscribers (Millions)")
plt.xlabel("")

for i, value in enumerate(df['Subscribers_Million']):
    plt.text(i, value+5, str(value), ha='center')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Company',
    y='Revenue_Billion_USD',
    data=df
)

plt.title("Netflix vs Amazon Revenue")
plt.ylabel("Revenue (Billion USD)")
plt.xlabel("")

for i, value in enumerate(df['Revenue_Billion_USD']):
    plt.text(i, value+10, str(value), ha='center')

plt.show()
```

```python
df['Revenue_per_Subscriber'] = (
    df['Revenue_Billion_USD'] * 1000
) / df['Subscribers_Million']

print(df[['Company','Revenue_per_Subscriber']])
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Company',
    y='Revenue_per_Subscriber',
    data=df
)

plt.title("Revenue Generated Per Subscriber")
plt.ylabel("Revenue per Subscriber (Approx USD)")
plt.xlabel("")

plt.show()
```

```python
import numpy as np

years = np.arange(2026, 2031)

netflix_subs = []
prime_subs = []

n = 302
p = 220

for year in years:
    netflix_subs.append(n)
    prime_subs.append(p)

    n *= 1.05
    p *= 1.08

forecast = pd.DataFrame({
    "Year": years,
    "Netflix": netflix_subs,
    "Amazon Prime": prime_subs
})

print(forecast)
```

```python
plt.figure(figsize=(10,6))

plt.plot(
    forecast['Year'],
    forecast['Netflix'],
    marker='o',
    label='Netflix'
)

plt.plot(
    forecast['Year'],
    forecast['Amazon Prime'],
    marker='o',
    label='Amazon Prime'
)

plt.title("Projected Subscriber Growth")
plt.ylabel("Subscribers (Millions)")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(10,6))

bars = plt.bar(
    ["Netflix","Amazon Prime"],
    [302,220]
)

plt.title(
    "Global Subscriber Comparison",
    fontsize=16,
    fontweight='bold'
)

plt.ylabel("Subscribers (Millions)")
plt.xlabel("")

for bar in bars:
    y = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        y + 5,
        f"{int(y)}M",
        ha='center',
        fontsize=12,
        fontweight='bold'
    )

plt.tight_layout()
plt.show()
```
