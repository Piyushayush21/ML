 ## A Confusion Matrix is used to evaluate a classification model.
  It compares:
  Actual values
 Predicted values

# TP (True Positive)
 Model predicted Positive and it is actually Positive.

Example:

Disease present
Model says disease present 

# TN (True Negative)

Model predicted Negative and it is actually Negative.

Example:

No disease
Model says no disease ✅

# FP (False Positive)

Model predicted Positive but actually Negative.

Example:

Healthy person predicted sick 

Also called:

Type 1 Error

 # FN (False Negative)

Model predicted Negative but actually Positive.

Example:

Sick person predicted healthy 

Also called:

Type 2 Error

# Precision

Precision tells:

Out of all predicted positives, how many were actually positive?

Formula:

Precision= TP+FP/TP
	​


Precision Example

Suppose:

Model predicted 100 people sick
Actually only 80 were sick

Then:

Precision=
80+20
80
	​

=0.8

Precision = 80%

# Recall

Recall tells:

Out of all actual positives, how many did model correctly identify?

Formula:

Recall=
TP+FN
TP
	​


Recall Example

Suppose:

Actually 100 sick people exist
Model identified only 90

Then:

Recall=
90+10/90
	​

=0.9

Recall = 90%

# Accuracy

Tells us Out of all predictions, how many predictions were correct?

It is mainly used in classification problems.

 # Formula of Accuracy

Accuracy=
TP+TN+FP+FN/TP+TN
	​


Where:

TP → True Positive
TN → True Negative
FP → False Positive
FN → False Negative
Simple Meaning

Accuracy = Total Predictions/Correct Predictions
	​


Correct predictions are:

TP
TN
